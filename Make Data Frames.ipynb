```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
filebase = 'data/mens-machine-learning-competition-2019/Stage2DataFiles/'

df = pd.read_csv(filebase + 'RegularSeasonDetailedResults.csv')

print(df.shape)
df.head()
```

```python
for c in df.columns:
    print(c)
```

```python
df_2 = pd.read_csv(filebase + 'NCAATourneyDetailedResults.csv')

print(df_2.shape)
df_2.head()
```

```python
for c in df_2.columns:
    print(c)
```

```python
df.columns == df_2.columns
```

```python
df = pd.concat([df, df_2])
df.shape
```

```python
common_columns = [i[1:] for i in df.columns if i.startswith('L')]
common_columns
```

```python
seasons = set(df['Season'])
seasons
```

```python
teams = set(df['WTeamID']).union(set(df['LTeamID']))
len(teams)
```

```python
n = len(seasons) * len(teams)
```

```python
new_cols = [f'AVG{c}' for c in common_columns[1:]]
new_cols
```

```python
new_cols.insert(0, 'Season')
new_cols.insert(0, 'TeamID')
new_cols.append('Wins')
new_cols.append('Losses')
new_cols
```

```python
avg_dict = {}

for c in new_cols:
    avg_dict[c] = [np.nan] * n
```

```python
iterrator = []

for team in teams:
    for season in seasons:
        iterrator.append((team, season))
        
len(iterrator) == n
```

```python
iterrator[0]
```

```python
for i, (team, season) in enumerate(iterrator):
    avg_dict['TeamID'][i] = team
    avg_dict['Season'][i] = season
    
    df_slice = df[(df['Season'] == season) & ((df['WTeamID'] == team) | (df['LTeamID'] == team))]
    
    for col in common_columns[1:]:
        col_total = 0
        col_count = 0
        
        for c in ['W', 'L']:
            try:
                series = df_slice[df_slice[f'{c}TeamID'] == team][f'{c}{col}']
                col_count += len(series)
                for val in series:
                    col_total += val
            except Exception as e:
                print(f'i: {i}\nteam: {team}\nseason: {season}\ncol: {col}\nc: {c}')
                raise e
        
        if col_count == 0:
            avg = 0
        else:
            avg = col_total / col_count
        
        avg_dict[f'AVG{col}'][i] = avg
    
    wins = len(df[(df['WTeamID'] == team) & (df['Season'] == season)])
    losses = len(df[(df['LTeamID'] == team) & (df['Season'] == season)])
    
    avg_dict['Wins'][i] = wins
    avg_dict['Losses'][i] = losses
```

```python
df_season_avg = pd.DataFrame.from_dict(avg_dict)
print(df_season_avg.shape)
df_season_avg.head()
```

```python
df_season_avg = df_season_avg[(df_season_avg['Wins'] > 0) | (df_season_avg['Losses'] > 0)]

df_season_avg.to_csv('season_avg.csv')
```

```python
avg_dict = {}
new_cols.remove('Season')
for c in new_cols:
    avg_dict[c] = [np.nan] * n
```

```python
for i, team in enumerate(teams):
    avg_dict['TeamID'][i] = team
    
    df_slice = df_season_avg[df_season_avg['TeamID'] == team]
    
    for col in new_cols:
        if col.startswith('AVG'):
            col_total = 0
            col_count = 0
            try:
                series = df_slice[col]
                col_count += len(series)
                for val in series:
                    col_total += val
            except Exception as e:
                print(f'i: {i}\nteam: {team}\ncol: {col}\nc: {c}')
                raise e
        
            if col_count == 0:
                avg = 0
            else:
                avg = col_total / col_count
        
            avg_dict[col][i] = avg
    
    wins = df_slice['Wins'].sum()
    losses = df_slice['Losses'].sum()
    
    avg_dict['Wins'][i] = wins
    avg_dict['Losses'][i] = losses
```

```python
df_avg = pd.DataFrame.from_dict(avg_dict)
print(df_avg.shape)
df_avg.head()
```

```python
df_avg.to_csv('avg.csv')
```

```python
columns = ['Team1', 'Team2'] + [f'Team1{c}' for c in common_columns[1:]]
columns += [f'Team2{c}' for c in common_columns[1:]] + ['Win']
columns
```

```python
n = len(df)
n
```

```python
matches_dict = {}

for c in columns:
    matches_dict[c] = [np.nan] * n
```

```python
import random

random.seed(42)

for i, row in df.iterrows():
    if i % 10000 == 0:
        print(f'{i}/{n}')
        
    t1 = random.choice(['W', 'L'])
    if t1 == 'W':
        t2 = 'L'
        win = 1
    else:
        t2 = 'W'
        win = 0
        
    team1 = row[f'{t1}TeamID']
    team2 = row[f'{t2}TeamID']
        
    team1_df = df_avg[df_avg['TeamID'] == team1]
    team2_df = df_avg[df_avg['TeamID'] == team2]
    
    matches_dict['Team1'][i] = team1
    matches_dict['Team2'][i] = team2
    matches_dict['Win'][i] = win
    
    for c in common_columns[1:]:
        matches_dict[f'Team1{c}'][i] = team1_df[f'AVG{c}'].item()
        matches_dict[f'Team2{c}'][i] = team2_df[f'AVG{c}'].item()
```

```python
matches_df = pd.DataFrame.from_dict(matches_dict)
print(matches_df.shape)
matches_df.head()
```

```python
matches_df.to_csv('matches.csv')
```
