```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('matches.csv')

print(df.shape)
df.head()
```

```python
#find unique item values with counts
for col in df.columns:
    print(col, "\n", max(df[col]), min(df[col]), sum(df[col])/df.shape[0], "\n\n")
```

```python
def replace_score(df, col_name):
    new_col = []
    for row in df[col_name]:
        if row >= 73.41:
            new_col.append(3)
        elif row >= 70.84:
            new_col.append(2)
        elif row >= 65.75:
            new_col.append(1)
        else:
            new_col.append(0)
    df[col_name] = new_col
    
replace_score(df, 'Team1Score')
replace_score(df, 'Team2Score')
print(df['Team1Score'].value_counts())
print(df['Team2Score'].value_counts())
```
