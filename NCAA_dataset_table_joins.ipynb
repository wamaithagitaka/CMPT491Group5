```python
col_map = {
    'WLoc': 'WLocation',
    'NumOT': 'NumOTPeriods',    
    'WFGM': 'WFieldGoalsMade',            
    'WFGA': 'WFieldGoalsAttempted',           
    'WFGM3': 'WThreePointersMade',
    'WFGA3': 'WThreePointersAttempted',
    'WFTM': 'WFreeThrowsMade',
    'WFTA': 'WFreeThrowsAttempted',
    'WOR': 'WOffensiveRebounds',
    'WDR': 'WDefensiveRebounds',
    'WAst': 'WAssists',
    'WTO': 'WTurnovers',   
    'WStl': 'WSteals',  
    'WBlk': 'WBlocks',    
    'WPF': 'WPersonalFouls',    
    'LFGM': 'LFieldGoalsMade',
    'LFGA': 'LFieldGoalsAttempted',
    'LFGM3': 'LThreePointersMade',
    'LFGA3': 'LThreePointersAttempted',
    'LFTM': 'LFreeThrowsMade',
    'LFTA': 'LFreeThrowsAttempted',
    'LOR': 'LOffensiveRebounds',
    'LDR': 'LDefensiveRebounds',
    'LAst': 'LAssists',
    'LTO': 'LTurnovers',   
    'LStl': 'LSteals',  
    'LBlk': 'LBlocks',    
    'LPF': 'LPersonalFouls'
}

columns = []

for column in df.columns:
    if column in col_map:
        columns.append(col_map[column])
    else:
        columns.append(column)

print(columns)

df.columns = columns
```

```python
print('Saving cleaned data')
# Save cleaned data
df.to_csv('data/ncaa_dataset_outer_join.csv', index=False)
```
